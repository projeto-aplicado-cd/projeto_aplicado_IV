# **Referencial Teórico**

Para o desenvolvimento do projeto de previsão de taxa de desemprego no Distrito Federal, nos basearemos em bibliografias que implementaram o modelo SARIMAX em conjuntos de dados de séries temporais e utilizaram variáveis consideradas fatores externos que influênciam os valores da série, com o objetivo de obter uma melhor acurácia e performance para as previsões do modelo. Nossa abordagem focará em compreender e aplicar conceitos estudados e testados anteriormente, nas referências mencionadas, para a solução de problemas similares.

O artigo “A Seasonal Autoregressive Integrated Moving Average with Exogenous Factors (SARIMAX) Forecasting Model-Based Time Series Approach” [Fahad Radhi Alharbi and Denes Csala 2022], utilizou o modelo SARIMAX para realizar previsões de 30 anos, num intervalo de 2021 a 2050, da performance do setor de energia na Arabia Saudita, combinando uma abordagem de séries temporais com sazonalidade e fatores de influência exógenos para redução de erro e melhoramento da acurácia.

A abordagem utilizada para a identificação dos melhores hiperparâmetros para o modelo foi o desenvolvimento de linhas de código em Python que fossem capaz de monitorar o desempenho e definir os melhores valores para (p, d, q) e (P, D, Q) a fim de obter a maior acurácia e divisão automática do conjunto de dados em 70% para treinamento e 30% para testes. Para a medição da acurácia foram utilizadas as métricas MAPE, RMSE, MAE, MSE e R2.

Comparações foram realizadas com outros modelos implementados em outros estudos e o SARIMAX proposto no artigo apresentou os melhores valores para a medição de erro e qualidade modelo. Os autores mencionam que “Os resultados do estudo também revelaram que o modelo SARIMAX superou seus concorrentes em termos de precisão de previsão, overfitting, eliminação de redundância, tempo de treinamento, e tempo de execução dos testes, comprovando que tem desempenho notável.” [Fahad Radhi Alharbi, Denes Csala, 2022, p.19].

Um outro trabalho usado como referência foi o “Métodos de previsão para a taxa de desemprego mensal: uma análise de séries temporais” [Dieison Lenon Casagrande, Felipe Resende Oliveira, Guilherme Studart, Inaldo Bezerra da Silva, Paulo Henrique Monteiro Guimarães 2016], que também realizou previsões utilizando o modelo SARIMAX com a inclusão da taxa de inflação como variável regressora, devido a uma possível relação de trade-off entre a taxa de desemprego e a taxa de inflação, ou seja, entende-se que uma inflação mais alta aquece a economia e gera uma maior procura por emprego, levando a uma menor taxa de desemprego.

Assim como o modelo do outro artigo mencionado neste tópico, o modelo desenvolvido para este segundo trabalho obteve um bom desempenho, apresentou o menor AIC - métrica usada para mensurar a qualidade de um modelo estatístico, quanto menor o valor do AIC, melhor o modelo é considerado em relação a outros modelos testados - em comparação ao modelo SARIMA, assim como, um menor erro de previsão, uma vez que apresentou um melhor ajuste.

Com base nas referências citadas, nota-se que modelos que incluem variáveis exógenas, para além de apenas utilizarem a própria série temporal, apresentam resultados melhores, erros de previsão baixos e um melhor ajuste aos dados. Portanto, para o nosso trabalho, focaremos em utilizar o modelo SARIMAX para realizar previsões de taxa de desemprego considerando os valores da taxa da Selic como variável externa e analisar a sua influência nas predições de desemprego.

Cálculo da taxa de desemprego segundo documentação da PED:

**PIA**: População em Idade Ativa: corresponde à população com quatorze anos e mais.

**PEA**: População Economicamente Ativa: corresponde à parcela da População em Idade Ativa (PIA) que está ocupada ou desempregada.

**Taxa de desemprego**: (Nº de desempregados PEA / Total PEA) x 100.

**Desemprego**: aberto + oculto (Desemprego Oculto pelo Trabalho Precário; Desemprego Aberto; Desemprego Oculto pelo Desalento)

```python
base_unica.situacao.unique()
```

```python
# filtra na base, os individuos com 14 anos ou mais e em situação de Ocupados ou Desempregados:
base_unica = base_unica[(base_unica.idade >= 14.0)]
base_unica = base_unica[base_unica.situacao.isin(['OCUPADO',
                                                  'OCUPADOS',
                                                  'DESEMPREGO ABERTO',
                                                  'DESEMPREGO OCULTO',
                                                  'DESEMPREGO OCULTO TRAB PREC',
                                                  'DESEMPREGO OCULTO PELO TRABALHO PRECÁRIO',
                                                  'DESEMPREGO OCULTO DESALENTO',
                                                  'DESEMPREGO OCULTO PELO DESALENTO'])]
```

```python
# verifica existencia de valores ausentes:
base_unica.isna().sum()
```

```python
# cria coluna com quantidade de pessoas que compoem a PEA (População Economicamente Ativa):
base_unica['count_pea'] = base_unica.groupby('data')['data'].transform('count')
```

```python
# cria coluna com quantidade de individuos em situacao de desemprego:
base_unica['count_des'] = base_unica[base_unica.situacao.isin(['DESEMPREGO ABERTO',
                                                                'DESEMPREGO OCULTO',
                                                                'DESEMPREGO OCULTO TRAB PREC',
                                                                'DESEMPREGO OCULTO PELO TRABALHO PRECÁRIO',
                                                                'DESEMPREGO OCULTO DESALENTO',
                                                                'DESEMPREGO OCULTO PELO DESALENTO'])].groupby('data')['data'].transform('count')
base_unica = base_unica[base_unica.count_des.notna()]
```

```python
# elimina duplicados:
base_unica = base_unica[['data','count_pea','count_des']].drop_duplicates()

# calcula taxa de desemprego:
base_unica['taxa_des'] = (base_unica['count_des']/base_unica['count_pea']) * 100
base_unica = base_unica[['data','taxa_des']].reset_index(drop=True)
```

```python
# verifica se todos os anos possuem 12 ocorrencias (12 meses):
base_unica.groupby(base_unica.data.dt.year).size()
```

Como podemos notar com a verificação da quantidade de registros por ano no dataframe, o ano de 2019 possui ocorrencia até Agosto (mês 8) e o ano de 2020 vai até Novembro (mês 11), portanto, preencheremos os demais meses faltantes através da técnica de interpolação.

```python
# Cria dataframe com os meses ausentes em 2019:
meses_ausentes = pd.to_datetime(['2019-09-01', '2019-10-01', '2019-11-01', '2019-12-01','2020-01-01'])
datas_ausentes = pd.DataFrame({'data': meses_ausentes, 'taxa_des': [None] * len(meses_ausentes)})

# Adiciona linhas para os meses ausentes com valores nulos:
base_unica = pd.concat([base_unica, datas_ausentes]).reset_index(drop=True)
base_unica['data'] = pd.to_datetime(base_unica['data'])
```

```python
# Preenche valores nulos para ano de 2019 e 2020, através da tecnica de interpolação linear:
base_unica.sort_values(by='data', ascending=True, inplace=True)
base_unica['taxa_des'] = base_unica['taxa_des'].interpolate(method='linear', limit_direction='backward')
```

```python
base_unica.sort_values(by='data', ascending=True, inplace=True)

# Plotting
plt.figure(figsize=(12,6))
plt.plot(base_unica['data'], base_unica['taxa_des'], label='Taxa de desemprego', color='green')
plt.legend()
plt.show()
```

### **Teste de Dickey-Fuller**

A seguir, realizamos o teste adfuller que verifica se a série é estacionária. Podemos interpretar os resultados como: se o valor do P-value for > 0.05, há um indicativo de NÃO estacionaridade da série temporal. Caso contrario, considera-se a série estacionária.

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# cria função que aplica o teste de adfuller para verificar estacionaridade da série:

def adfuller_test(df):
    dftest = adfuller(df, autolag = 'AIC')
    print("1. ADF : ",dftest[0])
    print("2. P-Value : ", dftest[1])
    print("3. Num Of Lags : ", dftest[2])
    print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
    print("5. Critical Values :")
    for key, val in dftest[4].items():
        print("\t",key, ": ", val)
```

```python
adfuller_test(final_df.taxa_des)
```

```python
adfuller_test(final_df.valor_selic)
```

```python
# tornando a série estacionaria atraves da diferenciação:
adfuller_test(final_df.taxa_des.diff().dropna())
```

Com os graficos abaixo, é interessante notar que a variavel **taxa_des** apresenta uma escala maior de sazonalidade do que a **valor_selic**, que apresenta uma escala muito pequena de 0.05.

Nesse trecho, nós utilizamos o **auto_arima** que é um algoritmo estatístico que automaticamente determina os melhores parametros para o nosso modelo. Definimos previamente alguns valores:


*   **m**: foi definido com base no ciclo de sazonalidade que ocorre a cada doze meses.
*   **max_p**: definimos como 2 o valor máximo do parametro p por causa dos resultados retornados e descritos na seção [Análise das correlações obtidas com a diferenciação da série](https://colab.research.google.com/drive/1UYdYFWGno5ug8xtIbIKNivUhHwy1Zoav#scrollTo=5QR3UNTp0Z5K) para o gráfico PACF.

*   **D**: definimos D = 0, pois não utilizamos uma diferenciação sazonal.





```python
# uso do auto_arima para descobrirmos quais os melhores parametros que retornam o menor AIC:
auto_arima(final_df['taxa_des'].dropna(),
           seasonal=True,
           m=12,
           max_p=2,
           d=1,
           D=0,
           trace=True,
           stepwise=True,
           ).summary()
```

## **Modelo XGBoost**

Nessa seção, empregamos o modelo de machine learning XGBoost para verificar o desempenho na predição da taxa de desemprego. O nosso objetivo e avaliar o desempenho desse modelo em comparação com os modelos estatísticos SARIMA e SARIMAX, empregados anteriormente.

### Criação das variáveis lag

A seguir, criamos um dataframe com os valores da taxa de desemprego + as colunas com as variáveis lag, que trazem os valores dessa taxa de desemprego para os 12 meses anteriores ao valor observado. Definimos 12 meses devido ao parametro m = 12 no nosso modelo SARIMA e SARIMAX, pois como foi observado na decomposição sazonal, o ciclo da sazonalidade se repete a cada 12 meses. Esses valores ajudam a capturar a dependência temporal dos dados.

```python
# definição de variáveis lag no nosso dataframe:
df_ml = pd.DataFrame()
n_lags = 12

for i in range(n_lags,0,-1):
    df_ml['taxa_des_' + 't-' + str(i)] = final_df[['taxa_des']].shift(i).values[:,0]
df_ml['taxa_des'] = final_df[['taxa_des']].values[:,0]
df_ml.index = final_df.index
df_ml = df_ml[n_lags:]

df_ml.head()
```

### Treinamento do modelo

```python
# define tamanho do conjunto de teste e treino:
test_size = int(len(df_ml) * 0.2)
train_size = len(df_ml) - test_size
```

```python
# separa as variaveis em dependente (y) e independente (X):
X = df_ml.drop(columns='taxa_des')
y = df_ml[['taxa_des']]

X_train, y_train  = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]
```

```python
from xgboost.sklearn import XGBRegressor

reg_xgb = XGBRegressor(n_estimators=1000, early_stopping_rounds=50)
reg_xgb.fit(X_train, y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
       verbose=25)
```

O XGBoost é um modelo baseado em árvore de decisão. No modelo acima, definimos os parametros n_estimators=1000 e early_stopping_rounds=50, para treinarmos 1000 árvores do modelo e pará-lo antecipadamente se não houver melhorias na métrica de avaliação RMSE após 50 interações.

### Predição e análise das métricas

```python
# predição sobre os dados de treino:
y_pred_xgb = reg_xgb.predict(X_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Plotar a série completa de dados reais
plt.figure(figsize=(10, 5))
plt.plot(y.index, y, label='Dados Reais (Completos)', color='black')

# Plotar os dados reais do conjunto de teste
plt.plot(y_test.index, y_test, label='Dados Reais (Teste)', color='gray')

# Plotar os dados previstos
plt.plot(y_test.index, y_pred_xgb, label='Dados Previstos', color='red', linestyle='--')

# Adicionar título e rótulos aos eixos
plt.title('Comparação entre Dados Reais (Completos), Dados Reais (Teste) e Dados Previstos')
plt.xlabel('Data')
plt.ylabel('Taxa de Desemprego')

# Adicionar legenda
plt.legend()

# Exibir o gráfico
plt.show()

# Calcular MSE
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
print(f"MSE: {mse_xgb}")

# Calcular RMSE
rmse_xgb = np.sqrt(mse_xgb)
print(f"RMSE: {rmse_xgb}")

# Calcular MAE
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
print(f"MAE: {mae_xgb}")
```

O que podemos observar analisando o gráfico e os valores é que as métricas obtidas apresentam valores próximos aos obtidos nos modelos SARIMA e SARIMAX, porém os modelos estatíscos desempenharam melhor. O resultado nos indica que se esse modelo XGboost for usado para forecasting, poderá apresentar um desvio de 1.34 unidades do valor real. Uma alternativa mais avançada para tentar melhorá-lo seria utilizar AUTO ML para a descoberta dos melhores parametros frente aos dados aplicados, ao invés do uso de valores padrões. Porém, nosso foco aqui, era fazer uma implementação mais **simplificada** para observar o **desempenho inicial** do modelo frente aos outros utilizados.

O artigo de Fahad Radhi Alharbi e Denes Csala (2022) é especialmente relevante, pois evidencia a superioridade do SARIMAX em termos de precisão e minimização de overfitting. De maneira semelhante, o estudo de Dieison Lenon Casagrande et al. (2016) corrobora essa abordagem, mostrando que a adição da taxa de inflação como variável exógena resultou em previsões mais precisas da taxa de desemprego.

Ao analisar os parâmetros para o nosso modelo SARIMAX, determinou-se que ( p = 0 ), ( d = 1 ), e ( q = 1 ), com componentes sazonais ( P = 1 ), ( D = 0 ), ( Q = 0 ) e sazonalidade anual (( m = 12 )). Esses parâmetros foram estabelecidos com base em análises de autocorrelação, autocorrelação parcial e utilização do algoritmo auto_arima para descoberta dos melhores parâmetros, garantindo que o modelo seja bem ajustado às características específicas dos dados.

# **Cronograma**

```python
#@title cronograma das etapas

import pandas as pd
import matplotlib.pyplot as plt

# Dados do cronograma
dados = {
    'Tarefa': ['Justificativa, Referencial Teórico e Cronograma', 'Implementação Parcial I', 'Implementação e Entrega Final'],
    'Inicio': ['2024-03-30', '2024-04-15', '2024-05-10'],
    'Fim': ['2024-04-07','2024-05-05','2024-05-29']
}

# Criar DataFrame a partir dos dados
df = pd.DataFrame(dados)

# Converter colunas de data para datetime
df['Inicio'] = pd.to_datetime(df['Inicio'])
df['Fim'] = pd.to_datetime(df['Fim'])

# Calcular a duração de cada tarefa
df['Duracao'] = df['Fim'] - df['Inicio']

# Plotar o cronograma
plt.figure(figsize=(10, 6))
for i, row in df.iterrows():
    plt.barh(y=row['Tarefa'], left=row['Inicio'], width=row['Duracao'], color='skyblue')
plt.xlabel('Data de Fim')
plt.ylabel('Tarefa')
plt.title('Cronograma de Tarefas')

# Definir os rótulos do eixo x como as datas de fim
plt.xticks(ticks=df['Fim'], labels=df['Fim'].dt.strftime('%Y-%m-%d'))

plt.grid(axis='x')
plt.show()
```
